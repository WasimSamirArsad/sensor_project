```python
# columns standard deviation is zero

def get_cols_with_zero_standard_dev(df:pd.DataFrame):
    """
    return the list of col having standard deviation is zero
    """
    cols_to_drop=[]
    num_cols=[col for col in df.columns if df[col].dtype != 'O']   # numerical cols
    for col in num_cols:
        if df[col].std()==0:
            cols_to_drop.append(col)
    return cols_to_drop
```

```python
get_cols_with_zero_standard_dev(wafers)
```

```python
def get_redundant_cols(df:pd.DataFrame,missing_threshold=.7):
    """
    return the col having missing value is greater than thresold
    """
    col_missing_ration=df.isna().sum().div(df.shape[0])
    cols_to_drop=list(col_missing_ration[col_missing_ration > missing_threshold].index)
    return cols_to_drop
```

```python
get_redundant_cols(wafers)
```

```python
# missing value more than 70%

cols_to_drop_1=get_redundant_cols(wafers,missing_threshold=0.7)
cols_to_drop_1
```

```python
# columns std = 0

cols_to_drop_2=get_cols_with_zero_standard_dev(wafers)
cols_to_drop_2.append("wafer")
cols_to_drop_2
```

## Data Transformation

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import KNNImputer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import RobustScaler
# from sklearn.preprocessing import StandardScaler

impute=KNNImputer(n_neighbors=3)
preprocessing_pipeline=Pipeline(steps=[("Imputer",impute),('scaler',RobustScaler())])
preprocessing_pipeline
```

```python
X_trans=preprocessing_pipeline.fit_transform(X)
print(X_trans.shape)
X_trans
```

```python
from imblearn.combine import SMOTETomek

X,y=X_trans[:,:-1],y
resampler=SMOTETomek(sampling_strategy="auto")
X_res,y_res=resampler.fit_resample(X,y)
```

```python
print("before",X.shape,y.shape)
print("after",X_res.shape,y_res.shape)
```

```python
# display score

def display_score(scores):
    print("scores  : ",scores)
    print("mean : ",scores.mean())
    print("std  : ",scores.std())
```
